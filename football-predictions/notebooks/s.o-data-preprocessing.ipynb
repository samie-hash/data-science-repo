### Convert columns to their appropriate types

```python
def date_to_regular_fmt(date_str: str):
    date_split = date_str.split('/')
    month = date_split[1]
    day = date_split[0]
    year = date_split[2]
    return f'{month}/{day}/{year}'
```

```python
date_to_regular_fmt('13/08/2015')
```

```python
football.head()
```

```python
football['Date'] = pd.to_datetime(football['Date'].apply(lambda x: date_to_regular_fmt(x)), infer_datetime_format=True)
```

```python
columns = ['FTHG','FTAG','HTHG','HTAG','HS','AS','HST','AST','HF','AF','HC','AC','HY','AY','HR','AR']
intconvert_tr = buif.IntegerConvert(columns=columns)
football = intconvert_tr.fit_transform(football)
```

### Add season label

```python
football['Year'] = football.Date.dt.year
```

```python
football.head()
```

```python
def remove_puntuations(x):
    return x.translate(str.maketrans('', '', string.punctuation))
```

```python
football.HomeTeam = football.HomeTeam.apply(lambda x: remove_puntuations(x))
football.AwayTeam = football.AwayTeam.apply(lambda x: remove_puntuations(x))
```
